```python
# Try to optimize: J(w1, w2) = w1**2 + w2**4
# Derivative = 2w1 + 4w2**3

# Formula w <-- w - 0.1*dj/dw
```

```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
w1 = 20
w2 = 25
X = list()
w1_hat = list()
w2_hat = list()

for i in range(100):
    X.append(i)
    w1 = w1 - 0.1*2*w1
    w1_hat.append(w1)

    w2 = w2 - 0.0003*4*w2**3
    w2_hat.append(w2)

w1_hat = np.array(w1_hat)
w2_hat = np.array(w2_hat)
w_hat = w1_hat + w2_hat
```

```python
plt.plot(X, w1_hat);
```

```python
plt.plot(X, w2_hat);
```

```python
plt.plot(X, w_hat);
```
